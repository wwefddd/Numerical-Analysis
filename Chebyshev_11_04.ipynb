cos 함수는 0부터 파이까지 감소한다. 그래서 xp에서 cos 값이 감소하는 순서로 잡았다. 그럼 xp는 -1 부터 1 까지 증가하는 순서가 된다

반원 위에서 같은 각으로 등분한 점들의 x좌표를 sampling 한다.
M은 자유롭게 대입할수 있지만, N은 하드코딩 된 상태여서 자유롭게 대입할 수 없다.

참고로 xp는 무조건 증가하는 순서여야 한다.
그리고 xp의 점의 개수에 따라서 분수 부분은 (1, 3, 5, 7, ... 2*점-1 )/(2 x 점) 형태이다.
예시) 아래 같은 경우에는 점이 4개 => 분수부분 : 1/8, 3/8, 5/8, 7/8 => cos은 감소함수이므로 순서 바꾸기
만약 점이 5개 => 분수부분 : 1/10, 3/10, 5/10, 7/10, 9/10 으로 하면 된다

```python
import numpy as np
import matplotlib.pyplot as plt

N = 4 # xp 점의 개수
M = 100 # 조사할 점의 개수
xp = np.zeros(N)
yp = np.zeros(N)
L = np.zeros((N, M))
Px = np.zeros(M)

x = np.linspace(-1, 1, M)

xp = np.array([np.cos(7/8 * np.pi), np.cos(3/8 * np.pi), np.cos(5/8 * np.pi), np.cos(1/8 * np.pi)])
yp = np.exp(xp)

for i in range(0, N): # 라그랑주 보간법
  L[i, :] = 1
  for j in range(0, i):
    L[i, :] = L[i, :] * (x - xp[j]) / ((xp[i] - xp[j]))
  for j in range(i+1, N):
    L[i, :] = L[i, :] * (x - xp[j]) / ((xp[i] - xp[j]))

for i in range(0, N):
  Px = Px + yp[i] * L[i, :]

max_err = np.max( np.abs(np.exp(x) - Px))
print( 'maximum error = ' , max_err)

plt.figure()
plt.plot(xp, yp, 'ro')
plt.plot(x, Px, '-')
plt.plot(x, np.exp(x), 'r:')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')

plt.figure()
plt.plot(x, np.abs(np.exp(x)-Px), '-')
plt.xlabel('x')
plt.ylabel('error')
plt.show()
```

일반화된 **버전**. N과 M을 자유롭게 대입할 수 있다.

```python
import numpy as np
import matplotlib.pyplot as plt

N = 20
M = 100
xp = np.zeros(N)
yp = np.zeros(N)
L = np.zeros((N, M))
Px = np.zeros(M)

x = np.linspace(-1, 1, M)

xp = np.zeros(N)
for j in range(0, N):
  xp[N - 1 - j]= np.cos((2 * j + 1)/ (2 * (N-1) + 2) * np.pi) # 2 * N +2 를 2 *(N-1) + 2 로 고칠것.
yp = np.exp(xp)

for i in range(0, N): # 라그랑주 보간법
  L[i, :] = 1
  for j in range(0, i):
    #print(i, j, xp[i], xp[j])
    L[i, :] = L[i, :] * (x - xp[j]) / ((xp[i] - xp[j]))
  for j in range(i+1, N):
    #print(i, j, xp[i], xp[j])
    L[i, :] = L[i, :] * (x - xp[j]) / ((xp[i] - xp[j]))

for i in range(0, N):
  Px = Px + yp[i] * L[i, :]

max_err = np.max( np.abs(np.exp(x) - Px))
print( 'maximum error = ' , max_err)

plt.figure()
plt.plot(xp, yp, 'ro')
plt.plot(x, Px, '-')
plt.plot(x, np.exp(x), 'r:')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')

plt.figure()
plt.plot(x, np.abs(np.exp(x)-Px), '-')
plt.xlabel('x')
plt.ylabel('error')
plt.show()
```
